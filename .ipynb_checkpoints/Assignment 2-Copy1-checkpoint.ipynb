```python
def xyz (query, api_key, rating, max_output):
    import string
    import random
    import requests
    import json
    url="https://maps.googleapis.com/maps/api/place/textsearch/json?query="
    query = query.replace(" ","+")
    url+= query
    url += "&key=" + api_key
    response = requests.get(url)
    if response.status_code != 200:
        print("not working")
    data_string = response.content.decode("utf-8")
    python_data = json.loads(data_string)
    m = python_data

    results_dict = m["results"]
    keep_keys = ["name","formatted_address","opening_hours", "price_level", "rating"]

    for business in results_dict:
        remove_keys=business.keys()-keep_keys
        for rkey in remove_keys:
            del business[rkey]

    mylist=[]
    for business in results_dict:  
        list1=[]
        for kkey in business:
            list1.append(business[kkey])
        mylist.append(tuple(list1))
        
    return mylist
```
